# Nederlandse Vlam Vereninging (NVV) Numerical Simulation Platform 

This document outlines ideas for a platform enabling the simuluation on large industrial furnaces in an energy transition (limit scope on type of furnaces targetted depending on interest of the partners?). This platform 
1. is intended as a means to share information;
2. is expected to act as a incubator for master and PhD projects;  
3. is expected to receive endorsement from the Nederlandse Vereniging voor Vlamonderzoek;  

Challenges on fuel side: fuel blends, hydrogen injection, staged injection, mild combustion; 

Challenges on oxidizer side: oxy-fuel combustion; 

The NVV board is aware that a single approach that accomodates all parties (one size fits all) is unlikely to exist. NVV recognize the adherence to public domain software as a strength to accomodate varies interests from a single common core.

## Section 1: Mission 

<b>Goals of a Common Simulation Platform</b>: 
To bridge the gap between sustainability (fuel reduction, fuel versatility, fuel staging (hydrogen, ammonia, syngas, CO2), oxy-fuel, flameless combustion, pollutant reduction and productivity (productivity closer to specs, design of new products and production processes). Gain understanding of how furnaces should be operated in new conditions (new fuel and/or now oxidizer conditions). Gain understanding of how new processes should be scaled up from lab to production environment; 

<b>Added Value for NVV Partners</b>: 
Contribute to competitive market advantage and social engagement of partners; 

<b>Definition of Program, Objectives and Study Cases</b>: 
The program will be shaped in dialogue with partners (Bekaert, Bosch, Celsian, Danieli-Corus, De Jong, Gouda Vuurvast, Origin Carbon Solution, Shell, XRG Technologies and others)

<b>Specific Deliverables</b>: 
Short and long term deliverables will be defined in dialogue with partners; 

<b>Scientific Challenges</b>: 
Dialogue with partners is expected create a pool of knowledge and to close the gap between existing and new knowledge. The use of public domain software (e.g. OpenFoam, SU2, PELE, ALYA and others) is expected to enable testing of established methods. 

## Section 2: Requirement for the Platform 

The platform will serve multiple partners on the long term. It therefore relies as much as possible on public domain license-free interchangeble software components.

### Simulation of heat generation 

Fuel and oxidizer mix, react and produce heat;  

### Simulation of heat absorption 

Heat generated diffuses and is generated by surrounding gas, fluid and solid; 

## Section 3: Choices for Physical Models 

### Models for Laminar/Turbulent Flow

Possible models for the laminar/turbulent flow of the mixture of fuel, oxidizer and combustion products include RANS, LES models, hybrid RANS/LES and (D)DES models. Requirements of mesh. Requirements on computational resources; 

### Models for Chemical Kinectics 

Depending on fuel composition and mode of combustion; 

### Models for Non/Partially Premixed Combustion 

Possible models for combustion include eddy-dissipation model (EDC (with or without the use of tabulation)), flamelet generated manifold (FGM), RIF and G-equation model; 

### Models for Radiative Heat Transfer in Participating Media 

Possible models for radiative heat transfer in participating media include spherical models, discrete ordinate models, Monte-Carlo ray tracing and others. Submodels are required to take presence of mixture of gasses, liquids or solids into account; 

### Models for Heat Transfer 
Models for conductive, convective and radiative heat transfer to gas, liquid or solid; 

## Section 4: Choices for Software 

The simulation platform is conceptually divided into the following parts: 
1. pre-processing step: geometry definition, mesh generation, flamelet table generation, parallel decomposition of inputs  and other; 
2. main computation step: solve reactive flow equations including combustion, radiative heat transfer and heat transfer to solid, liquid or gas phase; 
3. post-processing step: visualization of results 

The simulation platform can consist of commercial and public domain components. This components can be used side-by-side in such a way that information gained from various sources reinforces each other.  

### Commercial (Closed Source) Software 

<b>Advantages</b>: Commercial packages are robust (fail-save) and are equiped with a graphical user interface. They do remain very valuable as a tool for comparision.  

<b>Disvantages</b>: Commercial (off-the-shelf, closed source) software packages implement off-the-shelf models that might be ill-suited to tacke new needs in the energy transition. The use of use-defined functions allows some degree of customization. The public debate on this customization is hampered by the propriety nature of the software. The propriety nature of the software furthermore places its deployment behind closed doors (e.g. no hardware-specific (GPU) compilation, no linking with vendor-specific libraries, no profiling, no debugging) and limits its use on already existing and emerging hardware). To tackle new challenges, the shift to new tools might be required;

<b>Examples</b>: Examples include Ansys-Fluent (in use at e.g. Danieli-Corus, De Jong and XRG-Technologies) and ConvergeCFD;   
  
### Choice for OpenFoam 

<b>Advantages</b>: Extensive experience at TU Eindhoven, TU Delft (Mechnical Engineering, Math and Aerospace) and Shell (see [example](https://journal.openfoam.com/index.php/ofj/article/view/125)), the choice for OpenFoam is a natural choice. OpenFoam allows computations on parallel computing hardware without license costs;  

<b>Disvantages</b>: OpenFoam lacks components such as e.g. robust and computationally coupled pressure-velocity solver, steady-state reactive flow solver and supported flamelet solver including flamelet tables;

## Section 5: Choices for Hardware 

### Desktop computers 
Used in a first instance;  

### Parallel Computing hardware 

In later instances: parallel computers at SURFSARA, TU Delft or elsewhere. 

Commercial and public domain tools currently run of modern parallel computing hardware. The [ExaFoam project](https://exafoam.eu/challanges-and-goals/) is an example of limits are being pushed. Similar projects are currently developed for SU2 (see [SU2 2023 Users Meeting](https://su2foundation.org/su2conference2023/)) and for PELE (see e.g. https://amrex-combustion.github.io and https://www.exascaleproject.org/research-project/combustion-pele/);   

## Section 6: Validation of Simulation Models 

Validation of models for flow, combustion and heat transfer can be validated through a scale of test cases of increasing complexity. Easy to intermediate test cases can be taken from literature. Other intermediate and difficult test cases can be reproduced in a lab environment; 

## Section 7: Endorsement for the Platform 

### Danieli-Corus Ijmuiden 

[Danieli-Corus](https://www.danieli-corus.com) is an engineering firm with offices in Ijmuiden. It designs and manufactures equipment for iron and steel making. Danieli-Corus wishes to employ the simulation platform for the design of hot blast stoves (used to generate preheated combustion air) and blast stoves (used in the manufacturing of steel).

### Origen Carbon Solutions 
[Origen](https://www.origencarbonsolutions.com) is a company based in the UK active in direct air capture technology. Origen wishes to employ the NVV simulation platform to strengthen its suite of modelling tools to make informed decisions on scaling up production from pilot to industrial scale, among others.
