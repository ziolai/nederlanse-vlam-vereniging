
# Nederlandse Vlaam Vergenering

# Who we are - What is NVV

The NVV is a community of people, companies and institutions in the Netherlands which have interest in industrial research and education in the field of thermal systems ( combustion, gasification….) at lab, prototype or industrial scales. 

The NVV is a communication platform for close & fruitful contacts and collaborations between industry and academia.

Topics : Climate change, energy transition, sustainable industrial processes, sustainable transportation, energy storage, energy transportation, alternative fuels (e.g. biochar, ammonia)  and more...  

# Vision 

Establish a bridge between Industry and Accademia in the field of “combustion or thermal systems… experimental, theoretical and (semi-) industrial aspects ”, measurements techniques, computational fluid dynamics, reactive flows and more... 


# Objectives

Encourage consultation and cooperation between members, giving  members opportunities to meet for technical & scientific discussions & exchanges as well as professional education.

Foster collaboration among experts working at universities, technological institutes, industries or as independent professionals on a common interest.

Promote exchange of technical research results among members e.g. problem solving...

Prepare research programs [e.g NWO](https://www.nwo.nl/) and execute these programs in education and/or research entities with the collaboration of industrial members. 

Promote the full range of scientific knowledge on thermal systems and related topics and their dissemination and applications.

Encourage and initiate research in the European context through IFRF and the other committees [Links].

# Activities

Organisation of the Combura symposium in the fall of each year in cooperation with the Dutch Section of the Combustion Institute.

Combura is the place for the “combustion” community to meet and to exchange over industrial challenges & achievements and academic research & applications. [See Combura Page](https://www.aanmelder.nl/combura2025).

Organisation of the Spring Meeting typically on one of the members’ site where :

- The yearly members’ meeting takes place (&) Site visit

- All have time to network and exchange
